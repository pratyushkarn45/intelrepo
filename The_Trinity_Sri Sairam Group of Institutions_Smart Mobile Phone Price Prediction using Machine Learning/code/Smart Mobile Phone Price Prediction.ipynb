```python
from sklearn.metrics import confusion_matrix,classification_report
```

```python
def eval_model(ytest,ypred):
 cm = confusion_matrix(ytest,ypred)
 print(cm)
 print(classification_report(ytest,ypred))
def mscore(model):
 print('Train Score',model.score(x_train,y_train))
 print('Test Score',model.score(x_test,y_test))
```

# Trying Different Matric For Logistic Regression

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error

mse = mean_squared_error(y_test, ypred_lr1)
mae = mean_absolute_error(y_test, ypred_lr1)
print('Mean Squared Error:', mse)
print('Mean Absolute Error:', mae)
```

### Mean Absolute Error (MAE): MAE measures the average absolute difference between the predicted and actual prices. It provides a straightforward interpretation of the average prediction error and is less sensitive to outliers.

### Mean Squared Error (MSE): MSE calculates the average of the squared differences between the predicted and actual prices. It amplifies the impact of large errors and provides a more fine-grained evaluation of the model's performance.

```python
import numpy as np

# Assuming you have the actual prices in a numpy array `actual_prices` and the predicted prices in a numpy array `predicted_prices`

# Define the error threshold
error_threshold = 0.05  # Example: 5% error threshold

# Calculate the absolute error
absolute_error = np.abs(y_test - ypred_lr1)

# Count the number of predictions within the error threshold
within_threshold = np.sum(absolute_error <= error_threshold)

# Calculate the percentage of predictions within the error threshold
percentage_within_threshold = (within_threshold / len(y_test)) * 100

# Print the percentage within the error threshold
print("Percentage within", error_threshold * 100, "% error threshold:", percentage_within_threshold)
```

### This metric measures the percentage of predictions that fall within a certain error threshold (e.g., within 5% or 10% of the actual price). It provides insights into the accuracy of the model within a specific range of error.

```python
import numpy as np

# Assuming you have the actual values in a numpy array `actual_values` and the predicted values in a numpy array `predicted_values`

# Calculate the absolute error
absolute_error = np.abs(y_test - ypred_lr1)

# Calculate the median absolute error
medae = np.median(absolute_error)

# Print the MedAE value
print("Median Absolute Error (MedAE):", medae)
```

### MedAE measures the median absolute difference between the predicted and actual prices. It is less affected by outliers than MAE and can provide a more robust assessment of the model's performance.



```python
from sklearn.metrics import r2_score

# Assuming you have the actual values in a numpy array `actual_values` and the predicted values in a numpy array `predicted_values`

# Calculate the R-squared score
r2 = r2_score(y_test, ypred_lr1)

# Print the R-squared score
print("R-squared (R2) Score:", r2)
```

### The R-squared score represents the proportion of the variance in the dependent variable (actual values) that is predictable from the independent variable (predicted values) by the regression model. The score ranges from 0 to 1, where a score closer to 1 indicates a better fit of the model to the data. Note that a negative R-squared score can occur if the model performs worse than simply predicting the mean of the actual values.
